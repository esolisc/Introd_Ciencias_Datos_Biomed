### 2.1 Listas
Una lista guarda datos en orden. Los datos pueden ser de dititinto tipo.

```python
mi_lista = ["a", "b", "c", "d"]

print(mi_lista)
print(type(mi_lista))
print(mi_lista[0])    # 0: Primer elemento de la lista
print(mi_lista[2])    # 2: Tercer elemento de la lista
```

#### Modificar elementos de la lista

```python
mi_lista[0] = 'fabi'
print(mi_lista)
```

#### Añadir/quitar elementos de la lista

```python
mi_lista.append('último')
print(mi_lista)

mi_lista.pop()
print(mi_lista)
```

#### Acceder a varios elementos de la lista

```python
l = ['p','r','o','g','r','a','m','i','z']

print(l[2:5]) # Del 3o al 5o
print(l[:-5]) # Del principio al 4o
print(l[5:])  # Del 6o al final
print(l[:])   # Del principio al final
```

### 2.2 Diccionarios

Guarda pares de **clave: valor**

```python
d = { 'clave':   'valor',
     'nombre':   'Pedro',
     'apellido': 'García'}
```

Acceder al valor a través de la clave:

```python
print(d['clave'])
print(d['nombre'])
```

**Iterar** diccionario (forma 1):

```python
for clave in d:
    print("La clave es " + clave + " y el valor es " + d[clave])
```

**Iterar** diccionario (forma 2):

```python
for clave, valor in d.items():
    print("La clave es " + clave + ", y el valor es " + valor)
```

## 3. Control de flujo `if`, `elif`, `else`

```python
sueño  = True
hambre = False

if sueño and hambre:
    print("Come y a dormir la siesta.")
elif sueño and not hambre:
    print("A dormir la siesta")
elif hambre and not sueño:
    print("Come algo")
else:
    print("Disfruta del día")
```

## 4. Bucles (loops)
Sirven para **iterar** varios elementos.

### 4.1  Bucle `for`
Sirve para iterar un rangos, listas y diccionarios

```python
for i in range(10):
    print(i)
```

```python
range(10)
```

Iterar una lista

```python
animales = ['perro', 'gato', 'leon', 'oso']

for j in animales:
    print(j)
```

```python
for i, j in enumerate(animales):
    print(i, j)
```

Iterar un diccionario:

```python
sonidos_de_animales = {
    'perro': 'ladra',
    'gato': 'maulla',
    'león': 'ruge'
}

for animal, sonido in sonidos_de_animales.items():
    print("El " + animal + " " + sonido + ".")
```

### 4.2 Bucle `while`
Se usan para iterar con una condición.

```python
contador = 0
while (contador < 10):
    print(contador)
    contador += 1 # Incrementar el contador +1 . Es lo mismo que poner contador= contador + 1 (lo que ya venía + 1)
```

### 4.3 Lista de compresión

**Forma tradicional** de crear una lista: Empezar con una lista vacía y añadir elementos

```python
cuadrados = []
for n in range(10):
    cuadrados.append(n**2)
cuadrados
```

**List comprehensions**: Otra forma de crear listas más compacta.

```python
cuadrados = [n**2 for n in range(10)]
cuadrados
```

```python
abc_mayus = [letra.upper() for letra in 'abc' ]
print( abc_mayus)
```

### 4.4 Dict comprehensions

```python
cuadrados = {n: n**2 for n in range(10)}
cuadrados
```

## 5. Funciones

Una función es un bloque de código reutilizable. Una vez creada, se puede **llamar** en cualquier momento.

- Una funcion puede (o no) tener parametros de entrada.
- Una función puede (o no) devolver algo como salida.

Definir una función: (`def`)

```python
def es_primo(numero):
    if (numero==1):
        return False
    elif (numero==2):
        return True;
    else:
        for divisor in range(2,numero):
            if(numero%divisor == 0):
                return False
        return True
```

Llamar a la función. (Tambien se puede decir invocar o ejecutar)

```python
print("Es primo el 7?", es_primo(7))
print("Es primo el 8?", es_primo(8))
```

## Operaciones

```python
a = np.array([1,1, 1])
b = np.array([2, 3, 5])

print("a =", a)
print("b =", b)
print("==============")
print("a + b =", a + b)
print("b - a =", b - a)
print("a * b =", a * b) # Producto elementwise
print("b / a =", b / a)
print("a @ b =", a @ b) # Producto escalar (dot product)
print()
print("np.dot(a, b) =", np.dot(a, b)) # Producto escalar (dot product)
print("np.sqrt(b)   =", np.sqrt(b))   # Raiz cuadrada
print("np.mean(b)   =", np.mean(b))   # Media
print("np.std(b)    =", np.std(b))    # Desviacion estandar
print("np.median(b) =", np.median(b)) # Mediana
```

## Operaciones en una dimension determinada

```python
x = np.random.random((2, 8, 4))
x
```

```python
x.mean()
```

Si aplicamos la operacion sobre `axis=0`, la dimension 0 desaparace, (la forma final es de 8x4)

```python
x.mean(axis=0)
```

Si aplicamos la operacion sobre `axis=1`, la dimension 1 desaparace, (la forma final es de 2x4)

```python
x.mean(axis=1)
```

Si aplicamos la operacion sobre `axis=2`, la dimension 2 desaparace, (la forma final es de 2x8)

```python
x.mean(axis=2)
```

## Broadcasting
El Broadcasting nos permiter operar con arrays de distintas dimensiones. Por ejemplo imagina que a todos los numeros de un array los quieres multiplicar por 100.

Gracias al broadcasting podemos hacerlo:

```python
np.random.random((2,4)) * 100
```

Internamente, numpy convierte "el 100" en un array del mismo tamaño, y después computa la operación.
